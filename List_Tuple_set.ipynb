## Exercice sur les array ( Tableau numpy )
Créer un Tableu numpy array contenant les nombres entiers inférieur à 100 et qui sont des multiples de 3 et 7
> Tab= [ les multiples de 3 et de 7 qui sont <= 100]

```python
from numpy import array
T = array([int]*20)
# range(n) ={0,1,..,n-1 }
# range(a,b) = {a,a+1,...,b-1}
# range(a,b,s) = {a,a+s,a+2s,...,a+ks} where a+ks < b   
j = 0
for i in range(101):
    if i%3 == 0 and i%7 == 0:
        T[j] = i
        j += 1

print(T)
```
